# üõ§Ô∏è Ejercicio 1: Control de Tren (Valor - 6 pts)

### Descripci√≥n
Simula el sistema de control de un tren que debe responder a se√±ales en la v√≠a.  
Las se√±ales pueden ser **"verde"**, **"amarillo"** o **"rojo"**.  
El tren debe ajustar su velocidad base seg√∫n la se√±al actual y la distancia hasta la pr√≥xima estaci√≥n:

- **Se√±al verde**:  
  - Acelera a **100 km/h** si la distancia es mayor a 5 km.  
  - Mantiene **80 km/h** en caso contrario.  

- **Se√±al amarilla**:  
  - Reduce la velocidad a **50 km/h** si la distancia es mayor a 3 km.  
  - De lo contrario, reduce a **30 km/h**.  

- **Se√±al roja**:  
  - Frena hasta **0 km/h** inmediatamente.  

- **Obst√°culo en la v√≠a**:  
  - El tren debe frenar a **0 km/h**, independientemente de la se√±al.  

---

### ‚úÖ Salida esperada
El programa debe devolver la velocidad actual del tren en base a las condiciones dadas.  

**Ejemplo:**
- Entrada: `se√±al="amarillo", distancia=4, obst√°culo=False`  
- Salida: `Velocidad: 50 km/h`

```python
# Aqu√≠ se realiza el ejercicio 1
print("HOLAA hermano estoy aqui para ayudarte en las instruciones a seguir")
se√±al = input("Que se√±al se visualiza (roja , amarillo , verde )? : ").lower()

distancia = int(input("Cuantos km se encuentra de la estacion : "))

obstaculo = input("Ve algo que interumpe su circulacion (si/no) : ").lower()

#aqui hacemos solicitud de de los datos para realizar los comando para instruir al piloto

if se√±al == "verde" and distancia >= 5 and obstaculo == "no" :
    print("Acelere a 100km/h")  
    #se ve reflejado con los datos intruducidos como da la instrucion de aumentar la velocidad
elif se√±al == "verde" and distancia <= 4 and obstaculo == "no" :
    print("Mantenga velocidad de 80km/h")
    
elif se√±al == "amarillo" and distancia >= 3 and obstaculo == "no" :
    print("reduzca la velocidad a 50km/h")
    
elif se√±al == "amarillo" and distancia <= 3 and obstaculo == "no" :
    print("Reduzca la velocidad a 30km/h")
    
elif se√±al == "rojo" :
    print("Frene hasta los 0km/h")
    
elif obstaculo == "si" :
    print("Frene inmendiatamente")
    
else :
    print("introduzca bien los datos")
"""aqui vemos como ponemos las condiciones a los tipos de datos que no entran en este caso la distancia
, se√±al , obstaculo en la via y con estos datos podemos , dar las intrusciones a los pilotos"""
```
